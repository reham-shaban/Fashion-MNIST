# Classification using neural networks with keras

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

## Fashion-MNIST 
is a dataset of Zalando's article imagesâ€”consisting of a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes. 

```python
from PIL import Image
from IPython.display import display

img = Image.open('samples.png')
display(img)
```

### Load the data

```python
from keras.datasets import fashion_mnist

(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print(X_train_full.shape)
print(y_train_full.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
 "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
class_names[y_train_full[0]]
```

### Split into train and validation data
also scale the data to range 0-1 by dividing on 255

```python
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

### Creating the model

```python
model = Sequential([
    Flatten(input_shape=[28, 28]),
    Dense(300, activation="relu", name="hidden1"),
    Dense(100, activation="relu"),
    Dense(10, activation="softmax")
])
model.summary()
```

```python
weights, biases = model.get_layer("hidden1").get_weights()
print(weights.shape)
print(biases.shape)
```

### Compiling the Model

```python
model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="sgd",
    metrics=["accuracy"]
)
```

### Training and Evaluating the model

```python
history = model.fit(X_train, y_train, epochs=30,
                    validation_data=(X_valid, y_valid))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1) # set the vertical range to [0-1]
plt.show()
```

```python
history2 = model.fit(X_train, y_train, epochs=10,
                    validation_data=(X_valid, y_valid))
```

```python
model.evaluate(X_test, y_test)
```

### Using the model to make predictions
predict probapility

```python
X_new = X_test[:3]
y_proba = model.predict(X_new)
y_proba
```

predict class

```python
y_pred = np.argmax(y_proba, axis=1)
y_pred
```
